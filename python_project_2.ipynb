```python
import cv2
import numpy as np
```

```python
from matplotlib import pyplot as plt
car_cascade = cv2.CascadeClassifier('myhaar.xml')
```

```python
img = cv2.imread('image_3.png')
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
cars=car_cascade.detectMultiScale(gray,2.1,5)
```

```python
for (x,y,w,h) in cars:
  cv2.rectangle(img,(x,y),(x+w,y+h),(225,0,0),2)
```

```python
plt.imshow(img)
```
